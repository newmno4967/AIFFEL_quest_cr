```python
def find_min_max(numbers):
    # min_valueì™€ max_value ë³€ìˆ˜ë¥¼ ì´ˆê¸°í™”
    min_value = []
    max_value = []
    # min_valueëŠ” ì–‘ì˜ ë¬´í•œëŒ€(float('inf'))ë¡œ ì´ˆê¸°í™”í•˜ì—¬ ì–´ë–¤ ìˆ«ìë³´ë‹¤ë„ í° ê°’ìœ¼ë¡œ ì„¤ì •
    min_value = float('inf')
    # max_valueëŠ” ìŒì˜ ë¬´í•œëŒ€(float('-inf'))ë¡œ ì´ˆê¸°í™”í•˜ì—¬ ì–´ë–¤ ìˆ«ìë³´ë‹¤ë„ ì‘ì€ ê°’ìœ¼ë¡œ ì„¤ì •
    max_value = float('-inf')

    def update_min_max(num):
        # ì™¸ë¶€í•¨ìˆ˜ì˜ ë³€ìˆ˜ì¸ min_value, max_value ì°¸ì¡°
        nonlocal min_value, max_value
        # ë§Œì•½ num ê°’ì´ min_valueë³´ë‹¤ ì‘ë‹¤ë©´ min_valueë¥¼ num ê°’ìœ¼ë¡œ ë³€ê²½
        if num<min_value:
            min_value = num
        # ë§Œì•½ num ê°’ì´ max_valueë³´ë‹¤ í¬ë‹¤ë©´ max_valueë¥¼ num ê°’ìœ¼ë¡œ ë³€ê²½
        elif num>max_value:
            max_value = num

    # numbers ë¦¬ìŠ¤íŠ¸ì˜ ëª¨ë“  ê°’ì„ ìˆœí™˜í•˜ë©° ìµœëŒ“ê°’ê³¼ ìµœì†Ÿê°’ ì—…ë°ì´íŠ¸
    for num in numbers:
        update_min_max(num)

    # ìµœì†Ÿê°’ì„ ë°˜í™˜í•˜ëŠ” ë‚´ë¶€í•¨ìˆ˜
    def get_min():
        return min_value

    # ìµœëŒ“ê°’ì„ ë°˜í™˜í•˜ëŠ” ë‚´ë¶€í•¨ìˆ˜
    def get_max():
        return max_value

    # ì™¸ë¶€í•¨ìˆ˜ëŠ” ë‚´ë¶€í•¨ìˆ˜(get_min()ê³¼ get_max())ë¥¼ ë°˜í™˜
    return get_min, get_max
```

```python
numbers = [10, 5, 8, 12, 3, 7]
find_min, find_max = find_min_max(numbers)

print("ìµœì†Ÿê°’:", find_min())
print("ìµœëŒ“ê°’:", find_max())
```
