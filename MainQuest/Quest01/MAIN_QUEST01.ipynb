```python
from random import randrange # 랜덤함수 randrange만 불러오기

class Account: # Account클래스 생성
    user_account = [] # 생성되는 계좌정보를 기록할 리스트 생성

    # 이름과 초기잔액을 입력받아 객체 생성
    def __init__(self, name, first_money):
        bank_number = [] # 랜덤으로 생성될 계좌번호를 입력받을 변수 생성
        self.deposit_count = 0 # 이자를 받기위해 카운트할 변수 생성
        self.deposit_history_information = [] # 입금내역을 기록할 변수 생성
        self.withdraw_history_information = [] # 출금내역을 기록할 변수 생성
        self.name = name # 이름 입력
        self.money = int(first_money) # 초기자금 입력
        self.bank = 'SC은행' # 은행 입력
        # 랜덤으로 계좌번호 생성
        for i in range(11):
            bank_number.append(str(randrange(10)))
            if i in (2, 4):
                 bank_number.append('-')
        self.bank_account_number = ''.join(bank_number) # 계좌번호를 list형식에서 str형식으로 변환
        Account.user_account.append([self.name, self.bank_account_number, self.money]) # 생성된 계좌를 user_account에 저장
    
    # 총 생성된 계좌수를 확인하는 메서드(관계자만 확인해야 한다고 생각해서 이렇게 만들었습니다.)
    def get_account_num():
        print('총 생성된 계좌 개수: ', len(Account.user_account))
    
    # 입금 및 이자지급 메서드
    def deposit(self, money):
        if money <= 0: # 0원 입금시 안된다고 안내
            print('입금은 최소 1원 이상만 가능합니다.')
        else:
            self.money += money # 계좌에 돈 추가
            self.deposit_count += 1 # deposit_count를 1추가해 이자받기위한 준비
            self.deposit_history_information.append(['입금', money, self.money]) # deposit_history_information에 입금내역을 추가
            print('입금 후 금액:', self.money) # 입금후 금액 사용자에게 보여주기
        if self.deposit_count == 5: # deposit_count가 5를 넘을시
            self.money += self.money//100 # 1%의 이자 입금
            self.deposit_count = 0 # deposit_count초기화
            self.deposit_history_information.append(['이자지급', self.money//100, self.money]) # deposit_history_information에 이자입금 내역을 추가
            print('이자가 입금되었습니다. 추가 입금금액:', self.money//100, '총 입금 금액:', self.money) # 이자입금후 이자금액 및 총금액 사용자에게 보여주기
    
    # 입금 및 이자지급 내역 출력 메서드
    def deposit_history(self):
        for i in range(len(self.deposit_history_information)):
            print(f"{i+1}회: {self.deposit_history_information[i][0]}   금액: {self.deposit_history_information[i][1]}   잔액: {self.deposit_history_information[i][2]}")
    
    # 출금 메서드
    def withdraw(self, money):
        if self.money < money: # 계좌에 들어있는 돈보다 많은 돈을 출금하려할시 출력
            print('출금은 계좌의 잔고 이상으로 출금할 수는 없습니다.')
        else:
            self.money -= money # 계좌에서 출금
            self.withdraw_history_information.append(['출금', money, self.money]) # withdraw_history_information함수에 출금내역 기록
            print('출금 후 금액:', self.money) # 출금후 금액 출력
    
    # 출금 내역 출력 메서드
    def withdraw_history(self):
        for i in range(len(self.withdraw_history_information)):
            print(f"{i+1}회: {self.withdraw_history_information[i][0]}   금액: {self.withdraw_history_information[i][1]}   잔액: {self.withdraw_history_information[i][2]}")

    # 정보 출력 메서드
    def display_info(self):
        print(f"은행이름: {self.bank}, 예금주: {self.name}, 계좌번호: {self.bank_account_number}, 잔고: {self.money:,}")
    
    # 잔고가 100만원 이상인 고객만 출력하는 메서드
    def rich_user():
        print('잔고가 100만원 이상인 고객:')
        for i in range(len(Account.user_account)):
            if Account.user_account[i][2] >= 1000000:
                print(f"예금자명: {Account.user_account[i][0]}, 계좌번호: {Account.user_account[i][1]}, 잔고: {Account.user_account[i][2]}")


a = Account('a', 100)#
b = Account('b', 10000000)
c = Account('c', 6000)
d = Account('d', 999999999)
e = Account('e', 777777777)
f = Account('f', 1000)
Account.get_account_num()
a.display_info()
Account.rich_user()
print('-'*50)
a.deposit(100)
a.deposit(100)
a.deposit(100)
a.deposit(100)
a.deposit(100)
a.deposit(100)
a.deposit(100)
a.deposit(100)
print('-'*50)
a.withdraw(100)
a.withdraw(100)
a.withdraw(100)
a.withdraw(100)
a.withdraw(100)
a.withdraw(100)
print('-'*50)
a.deposit_history()
print('-'*50)
a.withdraw_history()
```

```python
from random import randrange # 랜덤함수 randrange만 불러오기

class Character: # Character클래스 생성
    # 기본값 생성
    level = 0
    strength = 0
    attack_power = 0
    defensive_power = 0
    def __init__(self):
        pass
    
    # 캐릭터가 생명력이 남아있는지 확인
    def  is_alive(self):
        if self.strength > 0:
            return True
    
    # self에게 damage_received만큼에 피해를 주기
    def take_damage(self, damage_received):
        if damage_received<self.defensive_power:
            pass
        else:
            self.strength -= damage_received-self.defensive_power
    
    # self의 데미지를 렌덤으로 생성
    def attack_target(self):
        self.damage = randrange(1, self.attack_power)
        return self.damage

class Player(Character): # Player클래스 생성
    # 이름을 입력받아 플레이어를 생성하고 기본값 설정
    def __init__(self):
        super().__init__()
        self.name = input("플레이어의 이름을 입력하시오 : ")
        self.level = 1
        self.strength = 100
        self.attack_power = 25 
        self.defensive_power = 5
        self.experience_quantity = 0 # 경험치 양

    # 경험치 획득 함수
    def gain_experience(self, experience):
        self.experience_quantity += experience
    
    # 경험치가 50이 넘으면 레벨업하는 함수
    def  level_up(self):
        if self.experience_quantity > 50:
            self.level += 1
            self.attack_power += 10
            self.defensive_power += 5
            self.experience_quantity -= 50
            print(self.name, "은 레벨업 하였다!")

class Monster(Character): # 몬스터 클래스 생성
    # 몬스터의 기본값 생성
    def __init__(self, name, level):
        super().__init__()
        self.name = name
        self.level = level
        self.strength = randrange(10, 30) * level
        self.attack_power = randrange(5, 20) * level
        self.defensive_power = randrange(1, 5) * level

def battle(player, monster): # battle함수 생성
    print(f"{monster.name}과의 전투를 시작합니다")
    while player.is_alive() and monster.is_alive(): # 몬스터와 풀레이어의 체력 둘중 1개라도 0이 될때까지 반복
        monster_damage = monster.attack_target() # monster_damage변수에 attack_target함수를 실행해 데미지 설정
        player_damage = player.attack_target() # player_damage변수에 attack_target함수를 실행해 데미지 설정
        monster.take_damage(player_damage) # take_damage함수를 실행해 플레이어가 몬스터 공격
        print(f"{player.name}이 {monster.name}에게 {player_damage}만큼 공격했다...!")
        print(f"{monster.name}의 체력: {monster.strength}")
        player.take_damage(monster_damage) # take_damage함수를 실행해 몬스터가 플레이어 공격
        print(f"{monster.name}이 {player.name}에게 {monster_damage}만큼 공격했다...!")
        print(f"{player.name}의 체력: {player.strength}")
    if Character.is_alive(player): # 반복문이 끝난 뒤 플레이어의 체력이 남아있으면 실행
        player.gain_experience(monster.level*20) # 경험치 획득
        player.level_up() # 레벨업
        print('전투 승리!')
        print(f"{monster.name}을 이겻다!")
    else: # 반복문이 끝난 뒤 플레이어의 체력이 없으면 실행 게임오버
        print('전투 패배...')
        print('게임오버')

def main(monster_list): # main함수 생성
    monster_dict = monster_list # 몬스터를 monster_dict에 기록

    monster = [] # 몬스터를 기록할 리스트 생성
    player = Player() # 플레이어 생성

    # monster_dict에 만들어둔 몬스터를 monster리스트에 넣기
    for monster_name in monster_dict:
        monster.append(Monster(monster_name, monster_dict[monster_name]))

    # 몬스터의 개수만큼 battle함수 실행
    for i in range(len(monster)):
        battle(player, monster[i])

main({'슬라임': 1, '고블린': 2, '오크': 3})
```
